### **Importing the libraries**

```
#importing the required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree
```

### **Importing the 'Iris' dataset into the notebook**

```
iris= "/content/drive/MyDrive/Iris.csv"
df= pd.read_csv(iris)
print(df.head()) #for first five rows
```

```
print (df.tail()) #for last five rows
```

```
df.info() #brief summary of DataFrame
```

```
df.describe()
```

```
df.isnull().sum()
```

### **Preparing the 'Iris' dataset**

```
df["Species"].unique()
```

```
x=pd.DataFrame(df[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]) #independent
y=df["Species"] #target
```

```
#dividing the dataset into train & test
from sklearn.model_selection import train_test_split
x_train , x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

### **Defining the Decision tree Algorithm**

```
#building the decision tree clasifier
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
print("Decision Tree Classifier Created!!")
```

```
y_pred = dt.predict(x_test)
```

```
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```
features = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm","PetalWidthCm"]
targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
```

### **Plotting/Visualization of the Decision Tree**

```
from sklearn.tree import plot_tree #visualizing the decision tree
from sklearn import tree
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(20,10), dpi=100)
tree.plot_tree(dt, feature_names=features, class_names=targets, filled=True)
fig.savefig('Decision_tree.png')
```
